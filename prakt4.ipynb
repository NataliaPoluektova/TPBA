**Завдання 1. Аналіз витрат на логістику**

Приклад: є дані про витрати на логістику по місяцях. Знайдіть середні витрати, місяці з перевищенням середнього, та побудуйте Series з відносною зміною.

Використовуємо

.mean() - обчислює середнє значення всіх елементів серії;

.pct_change() — обчислює відносну зміну згачення між поточним і попереднім у вигляді дробу (наприклад, 0.12 = +12%);

round(3) - округлення до трьох знаків після коми для зручності читання.

Виконайте код нижче:

```python
logistics = pd.Series(
    [8200, 8700, 7900, 9100, 9400, 9800],
    index=["Січ", "Лют", "Бер", "Квіт", "Трав", "Черв"],
    name="Логістика"
)
# Обчислюємо середнє значення витрат
mean_val = logistics.mean()
# Вибираємо місяці, де витрати перевищують середнє
above_avg = logistics[logistics > mean_val]
# Обчислюємо відносну зміну витрат між місяцями у відсотках
relative_change = logistics.pct_change().round(3)

print("Середнє:", mean_val)
print("Місяці з перевищенням:", above_avg)
print("Відносна зміна:\n", relative_change)
```

**Задача 3: Сезонність і тренди**

Умова: Проаналізуйте сезонність витрат — зсуви, різницю між місяцями, ковзне середнє.

.diff - обчислює різницю між поточним і попереднім елементом;

.rolling(3) - створює "вікно" з певної кількості елементів (наприклад, 3 місяці), і дозволяє застосовувати до нього статистичні методи: .mean(), .sum(), .std() тощо.
Виконайте приклад нижче:

```python
import pandas as pd

sales = pd.Series(
    [10000, 12000, 15000, 14000, 16000],
    index=["Січ", "Лют", "Бер", "Квіт", "Трав"],
    name="Продажі"
)

# Абсолютна зміна
abs_change = sales.diff()
print("Абсолютна зміна між місяцями:\n", abs_change)

# Відносна зміна (%)
pct_change = sales.pct_change().round(3) * 100
print("\nВідносна зміна (%):\n", pct_change)

# Ковзне середнє (вікно 3)
rolling_avg = sales.rolling(3).mean()
print("\nКовзне середнє (3 міс):\n", rolling_avg)
```

**Задача 5: Порівняння регіонів**
Умова: Знайдіть регіон з найбільшими витратами, частоту значень, унікальні суми.

.idxmax() повертає індекс (тобто назву регіону), де значення витрат є максимальним;

.value_counts() показує кількість повторень кожного значення;

.unique() повертає масив унікальних значень витрат.

Виконайте код нижче:

```python
regions = pd.Series(
    [12000, 13500, 12000, 14000, 13500, 15000],
    index=["Київ", "Львів", "Одеса", "Харків", "Дніпро", "Запоріжжя"]
)

print("Максимум у:", regions.idxmax())
print("Частота:\n", regions.value_counts())
print("Унікальні:\n", regions.unique())
```
