## –ú–µ—Ç–∞ —Ä–æ–±–æ—Ç–∏
1. –ó–∞—Å–≤–æ—ó—Ç–∏ —Ä–æ–±–æ—Ç—É –∑ Series —É –∫–æ–Ω—Ç–µ–∫—Å—Ç—ñ —Ä–µ–∞–ª—å–Ω–∏—Ö –±—ñ–∑–Ω–µ—Å-–∑–∞–¥–∞—á: –∞–Ω–∞–ª—ñ–∑ –≤–∏—Ç—Ä–∞—Ç, –¥–æ—Ö–æ–¥—ñ–≤, —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—ñ, –ø—Ä–∏–±—É—Ç–∫—É, —Ä–∏–∑–∏–∫—ñ–≤, –∑–≤—ñ—Ç–Ω–æ—Å—Ç—ñ.

–î–ª—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è Series –ø–æ—Ç—Ä—ñ–±–Ω–æ —ñ–º–ø–æ—Ä—Ç—É–≤–∞—Ç–∏ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É pandas:

```python
import pandas as pd
```

**–ó–∞–≤–¥–∞–Ω–Ω—è 1. –ê–Ω–∞–ª—ñ–∑ –≤–∏—Ç—Ä–∞—Ç –Ω–∞ –ª–æ–≥—ñ—Å—Ç–∏–∫—É**

–ü—Ä–∏–∫–ª–∞–¥: —î –¥–∞–Ω—ñ –ø—Ä–æ –≤–∏—Ç—Ä–∞—Ç–∏ –Ω–∞ –ª–æ–≥—ñ—Å—Ç–∏–∫—É –ø–æ –º—ñ—Å—è—Ü—è—Ö. –ó–Ω–∞–π–¥—ñ—Ç—å —Å–µ—Ä–µ–¥–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏, –º—ñ—Å—è—Ü—ñ –∑ –ø–µ—Ä–µ–≤–∏—â–µ–Ω–Ω—è–º —Å–µ—Ä–µ–¥–Ω—å–æ–≥–æ, —Ç–∞ –ø–æ–±—É–¥—É–π—Ç–µ Series –∑ –≤—ñ–¥–Ω–æ—Å–Ω–æ—é –∑–º—ñ–Ω–æ—é.

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ

.mean() - –æ–±—á–∏—Å–ª—é—î —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –≤—Å—ñ—Ö –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Å–µ—Ä—ñ—ó;

.pct_change() ‚Äî –æ–±—á–∏—Å–ª—é—î –≤—ñ–¥–Ω–æ—Å–Ω—É –∑–º—ñ–Ω—É –∑–≥–∞—á–µ–Ω–Ω—è –º—ñ–∂ –ø–æ—Ç–æ—á–Ω–∏–º —ñ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º —É –≤–∏–≥–ª—è–¥—ñ –¥—Ä–æ–±—É (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, 0.12 = +12%);

round(3) - –æ–∫—Ä—É–≥–ª–µ–Ω–Ω—è –¥–æ —Ç—Ä—å–æ—Ö –∑–Ω–∞–∫—ñ–≤ –ø—ñ—Å–ª—è –∫–æ–º–∏ –¥–ª—è –∑—Ä—É—á–Ω–æ—Å—Ç—ñ —á–∏—Ç–∞–Ω–Ω—è.

–í–∏–∫–æ–Ω–∞–π—Ç–µ –∫–æ–¥ –Ω–∏–∂—á–µ:

```python
logistics = pd.Series(
    [8200, 8700, 7900, 9100, 9400, 9800],
    index=["–°—ñ—á", "–õ—é—Ç", "–ë–µ—Ä", "–ö–≤—ñ—Ç", "–¢—Ä–∞–≤", "–ß–µ—Ä–≤"],
    name="–õ–æ–≥—ñ—Å—Ç–∏–∫–∞"
)
# –û–±—á–∏—Å–ª—é—î–º–æ —Å–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è –≤–∏—Ç—Ä–∞—Ç
mean_val = logistics.mean()
# –í–∏–±–∏—Ä–∞—î–º–æ –º—ñ—Å—è—Ü—ñ, –¥–µ –≤–∏—Ç—Ä–∞—Ç–∏ –ø–µ—Ä–µ–≤–∏—â—É—é—Ç—å —Å–µ—Ä–µ–¥–Ω—î
above_avg = logistics[logistics > mean_val]
# –û–±—á–∏—Å–ª—é—î–º–æ –≤—ñ–¥–Ω–æ—Å–Ω—É –∑–º—ñ–Ω—É –≤–∏—Ç—Ä–∞—Ç –º—ñ–∂ –º—ñ—Å—è—Ü—è–º–∏ —É –≤—ñ–¥—Å–æ—Ç–∫–∞—Ö
relative_change = logistics.pct_change().round(3)

print("–°–µ—Ä–µ–¥–Ω—î:", mean_val)
print("–ú—ñ—Å—è—Ü—ñ –∑ –ø–µ—Ä–µ–≤–∏—â–µ–Ω–Ω—è–º:", above_avg)
print("–í—ñ–¥–Ω–æ—Å–Ω–∞ –∑–º—ñ–Ω–∞:\n", relative_change)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ–≥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è:** –ê–Ω–∞–ª—ñ–∑ –≤–∏—Ç—Ä–∞—Ç –Ω–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥

–£ –≤–∞—Å —î Series –∑ –≤–∏—Ç—Ä–∞—Ç–∞–º–∏ –Ω–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥ –ø–æ –º—ñ—Å—è—Ü—è—Ö [5000, 5200, 5100, 5300, 5500, 5800]. –í–∏–∑–Ω–∞—á—Ç–µ, —á–∏ —î —Ç—Ä–µ–Ω–¥ –∑—Ä–æ—Å—Ç–∞–Ω–Ω—è. –Ø–∫—â–æ —Ç–∞–∫ ‚Äî –∑ —è–∫–æ—ó —Ç–æ—á–∫–∏ –≤—ñ–Ω –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è? –Ø–∫—ñ –º—ñ—Å—è—Ü—ñ –º–æ–∂–Ω–∞ –≤–≤–∞–∂–∞—Ç–∏ –Ω–µ—Å—Ç–∞–±—ñ–ª—å–Ω–∏–º–∏?

**–ó–∞–¥–∞—á–∞ 2: –ü—Ä–∏–±—É—Ç–æ–∫ —ñ –ø–æ–¥–∞—Ç–∫–∏**

–£–º–æ–≤–∞: —î –¥–æ—Ö–æ–¥–∏ —Ç–∞ –≤–∏—Ç—Ä–∞—Ç–∏. –û–±—á–∏—Å–ª—ñ—Ç—å –ø—Ä–∏–±—É—Ç–æ–∫, –ø–æ–¥–∞—Ç–∫–æ–≤–µ –Ω–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è (15%), —á–∏—Å—Ç–∏–π –ø—Ä–∏–±—É—Ç–æ–∫.

–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ:

```python
revenue = pd.Series([15000, 16000, 15500, 17000, 16500, 18000], index=logistics.index)
profit = revenue - logistics
tax = profit * 0.15
net_profit = profit - tax

print("–ü—Ä–∏–±—É—Ç–æ–∫:\n", profit)
print("–ü–æ–¥–∞—Ç–æ–∫:\n", tax)
print("–ß–∏—Å—Ç–∏–π –ø—Ä–∏–±—É—Ç–æ–∫:\n", net_profit)
```

–í–∞—à–∏ –≤–∏—Å–Ω–æ–≤–∫–∏ —Ç—É—Ç:

**–ó–∞–¥–∞—á–∞ 3: –°–µ–∑–æ–Ω–Ω—ñ—Å—Ç—å —ñ —Ç—Ä–µ–Ω–¥–∏**

–£–º–æ–≤–∞: –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ —Å–µ–∑–æ–Ω–Ω—ñ—Å—Ç—å –≤–∏—Ç—Ä–∞—Ç ‚Äî –∑—Å—É–≤–∏, —Ä—ñ–∑–Ω–∏—Ü—é –º—ñ–∂ –º—ñ—Å—è—Ü—è–º–∏, –∫–æ–≤–∑–Ω–µ —Å–µ—Ä–µ–¥–Ω—î.

.diff - –æ–±—á–∏—Å–ª—é—î —Ä—ñ–∑–Ω–∏—Ü—é –º—ñ–∂ –ø–æ—Ç–æ—á–Ω–∏–º —ñ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º –µ–ª–µ–º–µ–Ω—Ç–æ–º;

.rolling(3) - —Å—Ç–≤–æ—Ä—é—î "–≤—ñ–∫–Ω–æ" –∑ –ø–µ–≤–Ω–æ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, 3 –º—ñ—Å—è—Ü—ñ), —ñ –¥–æ–∑–≤–æ–ª—è—î –∑–∞—Å—Ç–æ—Å–æ–≤—É–≤–∞—Ç–∏ –¥–æ –Ω—å–æ–≥–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–Ω—ñ –º–µ—Ç–æ–¥–∏: .mean(), .sum(), .std() —Ç–æ—â–æ.
–í–∏–∫–æ–Ω–∞–π—Ç–µ –ø—Ä–∏–∫–ª–∞–¥ –Ω–∏–∂—á–µ:

```python
import pandas as pd

sales = pd.Series(
    [10000, 12000, 15000, 14000, 16000],
    index=["–°—ñ—á", "–õ—é—Ç", "–ë–µ—Ä", "–ö–≤—ñ—Ç", "–¢—Ä–∞–≤"],
    name="–ü—Ä–æ–¥–∞–∂—ñ"
)

# –ê–±—Å–æ–ª—é—Ç–Ω–∞ –∑–º—ñ–Ω–∞
abs_change = sales.diff()
print("–ê–±—Å–æ–ª—é—Ç–Ω–∞ –∑–º—ñ–Ω–∞ –º—ñ–∂ –º—ñ—Å—è—Ü—è–º–∏:\n", abs_change)

# –í—ñ–¥–Ω–æ—Å–Ω–∞ –∑–º—ñ–Ω–∞ (%)
pct_change = sales.pct_change().round(3) * 100
print("\n–í—ñ–¥–Ω–æ—Å–Ω–∞ –∑–º—ñ–Ω–∞ (%):\n", pct_change)

# –ö–æ–≤–∑–Ω–µ —Å–µ—Ä–µ–¥–Ω—î (–≤—ñ–∫–Ω–æ 3)
rolling_avg = sales.rolling(3).mean()
print("\n–ö–æ–≤–∑–Ω–µ —Å–µ—Ä–µ–¥–Ω—î (3 –º—ñ—Å):\n", rolling_avg)
```

**–ó–∞–≤–¥–∞–Ω–Ω—è –¥–ª—è —Å–∞–º–æ—Å—Ç—ñ–π–Ω–æ–≥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è:** –Ñ Series –∑ –≤–∏—Ç—Ä–∞—Ç–∞–º–∏ –Ω–∞ –ª–æ–≥—ñ—Å—Ç–∏–∫—É. –ü–æ–±—É–¥—É–π—Ç–µ –∫–æ–≤–∑–Ω–µ —Å–µ—Ä–µ–¥–Ω—î (–≤—ñ–∫–Ω–æ 3), –ø–æ—Ä—ñ–≤–Ω—è–π—Ç–µ –∑ —Ä–µ–∞–ª—å–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–Ω—è–º–∏. –í–∏–∑–Ω–∞—á—Ç–µ, —á–∏ —î —Å–µ–∑–æ–Ω–Ω—ñ –ø—ñ–∫–∏. –Ø–∫—ñ –º—ñ—Å—è—Ü—ñ –≤–∞—Ä—Ç–æ –æ–ø—Ç–∏–º—ñ–∑—É–≤–∞—Ç–∏?
